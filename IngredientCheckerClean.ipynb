```python
input_str = input("Enter ingredients to check (separated by commas): ")

input_list = [i.strip().lower() for i in input_str.split(",") if i.strip()]

comedogenic_lower = [i.lower() for i in comedogenic]
non_fungal_acne_safe_lower = [i.lower() for i in non_fungal_acne_safe]

comedogenic_found = []
non_fungal_acne_safe_found = []
safe_ingredients = []

for ing in input_list:
    if ing in comedogenic_lower:
        comedogenic_found.append(ing.title())
    elif ing in non_fungal_acne_safe_lower:
        non_fungal_acne_safe_found.append(ing.title())
    else:
        safe_ingredients.append(ing.title())

print("\n" + "="*50)
print("🔍 INGREDIENT ANALYSIS RESULTS")
print("="*50)

if comedogenic_found:
    print(f"\n❌ COMEDOGENIC ({len(comedogenic_found)} found):")
    for ing in comedogenic_found:
        print(f"   • {ing}")

if non_fungal_acne_safe_found:
    print(f"\n⚠️  NOT FUNGAL ACNE SAFE ({len(non_fungal_acne_safe_found)} found):")
    for ing in non_fungal_acne_safe_found:
        print(f"   • {ing}")

if safe_ingredients:
    print(f"\n✅ SAFE INGREDIENTS ({len(safe_ingredients)} found):")
    for ing in safe_ingredients:
        print(f"   • {ing}")

total_flagged = len(comedogenic_found) + len(non_fungal_acne_safe_found)
total_ingredients = len(input_list)

print(f"\n📊 SUMMARY:")
print(f"   Total ingredients checked: {total_ingredients}")
print(f"   Flagged ingredients: {total_flagged}")
print(f"   Safe ingredients: {len(safe_ingredients)}")

if total_flagged == 0:
    print("\n🎉 All ingredients are safe!")
else:
    print(f"\n⚠️  {total_flagged} ingredient(s) flagged for review.")
```
